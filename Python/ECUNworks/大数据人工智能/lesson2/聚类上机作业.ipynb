### 聚类上机作业
#### 1 利用numpy中的随机函数生成500个二维数据点，然后利用sklearn中的KMeans函数进行聚类分析，要求将聚类结果可视化表示出来。
#### 提示：使用numpy中的随机函数random.rand()，生成500个数据点

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn import neighbors
from sklearn import datasets
from sklearn.model_selection import cross_val_score
import math


Data=np.random.rand(500,2)
kmeans = KMeans(n_clusters=3).fit(Data)

# 获取聚类的标签
labels = kmeans.labels_

# 绘制数据点的散点图,通过c参数映射颜色到标签实现可视化
plt.scatter(Data[:,0], Data[:,1], c=labels)

# 绘制聚类中心,使用黑色的X表示
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], marker='x', s=100, c='red') 

# 添加标题  
plt.title('KMeans聚类')
plt.show()

```

#### 2 利用sklearn中的make_bolbs()函数随机生成200个样本数据，样本数据特征数为2个，然后利用KMeans函数进行聚类分析，并显示结果。

```python
from sklearn.datasets import make_blobs 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 生成200个样本,每个样本2个特征  
X, y = make_blobs(n_samples=200, n_features=2, centers=3)

# KMeans聚类 
kmeans = KMeans(n_clusters=3)  
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 绘制散点图
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50) 

# 绘制聚类中心
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.5)

# 添加标题
plt.title('KMeans on Blobs')
plt.show()
```

#### 3 对sklearn中的鸢尾花数据进行聚类分析，聚类簇数可以设置为2至5，并通过可视化聚类结果的办法，观察聚类簇数选择多少效果比较好。

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 加载数据集
iris = load_iris()
X = iris.data  

# 尝试不同的聚类数
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k).fit(X)
    y_kmeans = kmeans.predict(X)
    
    # 绘制散点图  
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50)

    # 绘制聚类中心
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

    # 标题
    plt.title('K = %d' % k)
    plt.show()
```
