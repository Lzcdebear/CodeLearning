## 5.4.3 K-Means综合实践
&emsp;&emsp;本节通过三个从简单到复杂的例子来说明scikit-learn中KMeans算法的应用。  
<br>
&emsp;&emsp;**例5-4-1** 利用scikit-learn的KMeans算法实现5.4.3节中的例子。

#### （1）导入库
&emsp;&emsp;通过import导入numpy库用于生成样本，导入matplotlib库用于可视化图表，导入sklearn.cluster用于聚类。

```python
#（1）导入库
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

#### （2）生成样本数据
&emsp;&emsp;使用numpy库生成包含示例数据的6个样本数据，每个样本都是两维的。

```python
#（2）生成样本数据
samples = np.array([[1,1],[2,2],[3,1],[6,4],[7,5],[8,4]])
```

#### （3）把样本数据显示在二维坐标上
&emsp;&emsp;语句plt.figure(figsize=(8,5),dpi=144)，创建画布，并设置图形的大小为8\*5英寸，设置图形每英寸的点数为144个。  

&emsp;&emsp;语句plt.scatter(samples\[:,0],samples\[:,1],s=100)，以samples数据集的第0列和第1列数据作为两个维度，绘制散点图，并设置数据点标记大小为100。

```python
#（3）把样本数据显示在二维坐标上
plt.figure(figsize=(8,5),dpi=144)
plt.scatter(samples[:,0],samples[:,1],s=100)
```

#### （4）聚类训练
&emsp;&emsp;使用KMeans模型来拟合，设置聚类簇数为2，并使用fit()函数进行训练。

```python
#（4）使用KMeans模型拟合
est=KMeans(n_clusters=2)
est.fit(samples)
```

#### （5）展示结果
&emsp;&emsp;绘制聚类结果和质心。 

&emsp;&emsp;语句plt.scatter(samples\[:,0],samples\[:,1],s=100,c=labels.astype(np.float))，以样本集samples为数据绘制散点图，设置数据点大小为100，同时由于已经完成了聚类，数据点已具有类别，因此设置数据点颜色为其对应的标签值。  

&emsp;&emsp;语句plt.scatter(centers\[:,0],centers\[:,1],s=100,marker="\*")，以质心为数据绘制散点图，并设置数据点大小为100，形状为*。  

&emsp;&emsp;结果如下图所示，可以看出六个样本点被聚成两类，\*表示质心。

```python
#（5）将聚类结果利用散点图显示出来
labels=est.labels_
centers=est.cluster_centers_
fig=plt.figure(figsize=(8,5),dpi=144)
plt.scatter(samples[:,0],samples[:,1],s=100,c=labels.astype(np.float))
plt.scatter(centers[:,0],centers[:,1],s=100,marker="*")
plt.show()
```

&emsp;&emsp;**例5-4-2** 利用随机生成的一组样本数据进行聚类分析。

#### （1）导入库 
&emsp;&emsp;sklearn.datasets.samples_generator用于生成样本数据，matplotlib用于可视化图表，sklearn.cluster用于聚类。

```python
#（1）导入库
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

#### （2）生成样本数据集
&emsp;&emsp;语句X,Y=make_blobs(n_samples=1000,centers=2)，生成1000个样本、2个中心点的样本数据集，并将生成的样本返回给X变量，标签返回给Y变量。

```python
#（2）利用scikit-learn中的make_blobs函数生成样本数据集
X,Y=make_blobs(n_samples=1000,centers=2)
```

#### （3）把样本数据显示在二维坐标上
&emsp;&emsp;绘制散点图将样本数据显示在二维坐标上，以便直观地观察。 

&emsp;&emsp;语句plt.figure(figsize=(8,5),dpi=144)，创建画布，设置图形的大小为8\*5英寸，设置图形每英寸的点数为144个。  

&emsp;&emsp;语句plt.scatter(X\[:,0],X\[:,1],s=50,edgecolor='k')，以X数据集的第0列和第1列数据作为两个维度，绘制散点图，设置数据点标记大小为50，同时为便于显示数据点设置数据点边缘为黑色。

```python
#（3）利用散点图的形式将样本数据展示出来
plt.figure(figsize=(8,5),dpi=144)
plt.scatter(X[:,0],X[:,1],s=50,edgecolor='k')
plt.show()
```

#### （4）聚类训练
&emsp;&emsp;使用KMeans模型来拟合，设置聚类簇数为4，并使用fit()函数训练模型。

```python
#（4）使用KMeans模型拟合，聚类数设为4。
kmean=KMeans(4)
kmean.fit(X)
```

#### （5）展示结果
&emsp;&emsp;将聚类结果和质心利用散点图显示出来。 

&emsp;&emsp;语句plt.scatter(X\[:,0],X\[:,1],c=labels.astype(int),s=50,edgecolor='k')，以X数据集的第0列和第1列数据作为两个维度，绘制散点图，根据聚类类别设置数据点的颜色，设置数据点标记大小为50，设置数据点边缘为黑色。  

&emsp;&emsp;语句plt.scatter(centers\[:,0],centers\[:,1],c='r',s=100,marker="\*")，以质心为数据绘制散点图，并设置颜色为红色，大小为100，形状为*。  

&emsp;&emsp;结果如下图所示，可以看出样本点被聚成4类，\*表示质心。

```python
#（5）将聚类结果利用散点图显示出来
labels=kmean.labels_
centers=kmean.cluster_centers_
fig=plt.figure(figsize=(8,5),dpi=144)
plt.scatter(X[:,0],X[:,1],c=labels.astype(int),s=50,edgecolor='k')#显示聚类结果
plt.scatter(centers[:,0],centers[:,1],c='r',s=100,marker="*")#显示质心
plt.show()
```

&emsp;&emsp;**例5-4-3** 利用scikit-learn的K-Means算法实现一个高维数据的聚类操作。该例的数据来自scikit-learn中的鸢尾花数据Iris。

#### （1）导入库
&emsp;&emsp;sklearn.datasets用于导入数据加载器，matplotlib和mpl_toolkits.mplot3d用于可视化图表，sklearn.cluster用于聚类。然后，设置plt.rcParams\['font.sans-serif']为\['SimHei']，避免中文出现乱码，为后续画图做好准备。

```python
#（1）导入库
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
plt.rcParams['font.sans-serif']=['SimHei'] #避免中文出现乱码
```

#### （2）导入数据集
&emsp;&emsp;使用加载器读取数据并存入数据集变量iris，然后将iris数据集的特征数据存入变量X。

```python
#（2）导入Iris数据
iris=datasets.load_iris()#导入iris数据
X=iris.data
```

#### （3）聚类训练
&emsp;&emsp;使用KMeans模型来拟合，设置聚类簇数为3，并使用fit()函数训练模型。

```python
#（3）使用KMeans模型拟合，聚类数设为3
est=KMeans(n_clusters=3)
est.fit(X)
```

#### （4）展示结果
&emsp;&emsp;将聚类结果利用散点图显示出来。

&emsp;&emsp;语句ax=Axes3D(fig,elev=48,azim=134)，创建Axes3D对象ax，设置当前的图像为fig，设置仰角为48度，设置方位角为134度。 

&emsp;&emsp;语句ax.scatter(X\[:,3],X\[:,0],X\[:,2],c=labels.astype(np.float),edgecolor='k')，以X数据集的第3列、第0列和第2列数据（即花萼宽度、萼片长度和花瓣长度）作为坐标，绘制散点图，根据聚类类别设置数据点的颜色，设置数据点边缘为黑色。  

&emsp;&emsp;语句ax.set_xlabel('花萼宽度')、ax.set_ylabel('萼片长度')、ax.set_zlabel('花瓣长度')，分别设置三个坐标轴的名称。  

&emsp;&emsp;语句ax.set_title('Iris数据集的聚类展示')，设置了整个散点图的名称为“'Iris数据集的聚类展示”。  

&emsp;&emsp;语句ax.dist=12，设置了视角的观看距离。调整这个值，可以放大缩小图片。

```python
#（4）选取其中的三个维度，并显示其聚类结果
labels=est.labels_
fig=plt.figure(figsize=(8,5),dpi=144) 
ax=Axes3D(fig,elev=48,azim=134)
ax.scatter(X[:,3],X[:,0],X[:,2],c=labels.astype(np.float),edgecolor='k')
ax.set_xlabel('花萼宽度')
ax.set_ylabel('萼片长度')
ax.set_zlabel('花瓣长度')
ax.set_title('Iris数据集的聚类展示')
ax.dist=12
plt.show()
```
